```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:\\Users\\sayak\\anaconda3\\noisyimg.png',0)
new_img3 = cv2.imread('C:\\Users\\sayak\\anaconda3\\noisyimg.png', 0)
new_img5 = cv2.imread('C:\\Users\\sayak\\anaconda3\\noisyimg.png', 0)
prop = img.shape
```

```python
############### 3x3 window ############### 
for i in range(1, prop[0] - 1):
    for j in range(1, prop[1] - 1):
        win = []
        for x in range(i-1, i + 2):
            for y in range(j-1, j+2):
                win.append( img[x][y] )
        #sort the values
        win.sort()

        new_img3[i][j] = win[4]
```

```python
############### 5x5 window ###############
for i in range(1, prop[0] - 2):
    for j in range(1, prop[1] - 2):
        win = []
        for x in range(i - 2, i + 3):
            for y in range(j - 2, j + 3):
                win.append(img[x][y])
        #sort the values
        win.sort()

        new_img5[i][j] = win[12]
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(img, 'gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(new_img3, 'gray')
plt.title('3x3')
plt.subplot(1,3,3)
plt.imshow(new_img5, 'gray')
plt.title('5x5')
```
